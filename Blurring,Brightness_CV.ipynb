```
# Python Program to blur image 
  
# Importing cv2 module 
import cv2  
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/My Drive/person.jpg')  
plt.imshow(img) 
```

```
# average of all the pixels under kernel area
blurImg = cv2.blur(img,(10,10))  
plt.imshow(blurImg) 
```

```
# Gaussian Blurring 
# instead of a box filter consisting of equal filter coefficients, a Gaussian kernel is used
IMG_DIM = 299
img = cv2.resize(img, (IMG_DIM, IMG_DIM))
gausBlur = cv2.GaussianBlur(img, (0,0) , IMG_DIM/10)  
plt.imshow(gausBlur) 
```

```
img1=cv2.addWeighted ( img,4, gausBlur,-4 ,128)
plt.imshow(img1)
```

```
# Median blurring 
#highly effective in removing salt-and-pepper noise.
medBlur = cv2.medianBlur(img,5) 
plt.imshow(medBlur) 
```

```
# Bilateral Filtering 
#Gaussian component applied in the intensity domain (a Gaussian function of intensity differences) ensures that only those pixels with intensities similar to that of the central pixel (‘intensity neighbors’) are included to compute the blurred intensity value
bilFilter = cv2.bilateralFilter(img,9,75,75) 
plt.imshow(bilFilter)
```

```
# Python program for blending of 
# images using OpenCV 
  
# import OpenCV file 
import cv2 
  
# Read Image1 
mountain = cv2.imread('/content/drive/My Drive/coins.jpg', 1) 
mountain_image = cv2.resize(mountain,(100,100))   
# Read image2 
dog = cv2.imread('/content/drive/My Drive/car.jpg', 1) 
dog_image = cv2.resize(dog,(100,100)) 
  
# Blending the images with 0.3 and 0.7 
img = cv2.addWeighted(mountain_image , 0.3, dog_image, 0.7, 0) 
  
# Show the image 
plt.imshow(img) 
```
